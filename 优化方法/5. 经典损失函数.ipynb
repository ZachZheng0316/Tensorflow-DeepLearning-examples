```python
import tensorflow as tf
import numpy as np
from numpy.random import RandomState
import matplotlib.pyplot as plt
```

```python
## 创建测试数据
rdm = RandomState(1)
data = [np.sin(x/100.) for x in range(107)]
data_Pre = [np.sin(x/100.)+(rdm.rand()/10.0-0.05) for x in range(107)]
Train_Y = tf.constant(value=data)
Pre_Y = tf.constant(value=data_Pre)

plt.plot(np.arange(len(data)), data, label="data")
plt.plot(np.arange(len(data)), data_Pre, label="data_Pre")
plt.xlabel("Time(d)")
plt.ylabel("Sales")
plt.legend()
plt.show()
```

```python
## MASE:绝对值误差
```

$$
J = \frac{1}{m}\sum_{i=1}^{m}\left \| y - y^p \right \|
$$

```python
loss = tf.losses.absolute_difference(data, data_Pre)
print("loss: {}".format(loss))
```

# 问题描述

在sin(x) 函数上, 用前 TIMESTEPS 个点(数据)预测第 TIMESTEPS + 1 个数据.

# 收集数据

用正弦函数sin生成训练以及测试的数据集合.

先生成一个训练数据和测试数据集合, 例如:
    
- 训练数据集合:[sin(0.01), sin(0.02), ..., sin(100.00)]: 10000个数据, x轴数据间隔为0.01
- 测试数据集合:[sin(100.01), sin(100.02), ..., sin(110.00)]: 1000个数据, x轴数据间隔为0.01

```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
TRAINING_EXAMPLES = 10000 # 训练数据的个数
TESTING_EXAMPLES = 1000 # 测试数据的个数
SAMPLE_GAP = 0.01 # 采样间隔
TIMESTEPS = 10 # 神经网络的训练长度
```

```python
test_start = (TRAINING_EXAMPLES + TIMESTEPS) * SAMPLE_GAP
test_end = test_start + (TESTING_EXAMPLES + TIMESTEPS) * SAMPLE_GAP

# 生成训练集原始数据集合
training_data = np.sin(np.linspace(0, test_start, TRAINING_EXAMPLES + TIMESTEPS, dtype=np.float32))

# 生成测试集原始数据集合
test_data = np.sin(np.linspace(test_start, test_end, TESTING_EXAMPLES + TIMESTEPS, dtype=np.float32))
```

# 整理数据

把数据整理成如下形式:
    
- feature:[0.00, 0.01, ... , 0.09]
- label:[0.10]

训练集:

- training_set:[feature1, ..., feature10000]
- training_label:[label1, ..., label10000]

测试集数据形式类似

```python
# 生成feature与label的组合
def generate_data(seq):
    X = []
    label = []
    
    # 计算feature和label
    for i in range(len(seq) - TIMESTEPS):
        # 添加第i个特征数据
        X.append([seq[i: i + TIMESTEPS]])
        # 添加第i个label
        label.append([seq[i + TIMESTEPS]])
        
    # 返回创建的列表
    return np.array(X, dtype=np.float32), np.array(label, dtype=np.float32)

# 产生训练数据
train_X, train_label = generate_data(training_data)
train_X = train_X.reshape(TRAINING_EXAMPLES, TIMESTEPS, -1)
print("train_X shape: ", train_X.shape)
print("train_label: ", train_label.shape)

# 产生测试数据
test_X, test_label = generate_data(test_data)
test_X = test_X.reshape(TESTING_EXAMPLES, TIMESTEPS, -1)
```

# 构建训练模型

模型结果如下:
    
![sin预测](sin预测.png)



```python
HIDDEN_SIZE = 30
NUM_LAYERS = 2

# 定义模型结构
def lstm_model(X, label, is_training):
    # 使用多层lstm结构
    lstm_cell = [tf.nn.rnn_cell.BasicLSTMCell(HIDDEN_SIZE) for _ in range(NUM_LAYERS)]
    cell = tf.nn.rnn_cell.MultiRNNCell(lstm_cell)
    
    # 使用TensorFlow接口将多层的LSTM结构连接成RNN网络并计算其前向传播解雇
    # X shape: [batch, TIMESTEPS, Embedding_size]
    # outputs shape: [batch, TIMESTEPS, HIDDEN_SIZE]
    outputs, _ = tf.nn.dynamic_rnn(cell, X, dtype=tf.float32)
    # 本问题只关心最后一个时刻的输出结果, 该结果为下一个时刻的预测值
    output = outputs[:, -1, :] # 获取最后一个输出
    print("outputs shape: ", tf.shape(outputs))
    
    # 对LSTM网络的输出再做加一层全链接层并计算损失,输出节点为1个。
    # 注意这里默认的损失为平均平方差损失函数。
    # predictions为预测值
    predictions = tf.contrib.layers.fully_connected(
        output, 1, activation_fn=None)
    
    # 只在训练时计算损失函数和优化步骤。测试时直接返回预测结果。
    if not is_training:
        return predictions, None, None
    
    # 计算损失函数
    loss = tf.losses.mean_squared_error(labels=label, predictions=predictions)
    
    # 创建模型优化器并得到优化步骤
    train_op = tf.contrib.layers.optimize_loss(loss, 
                                               tf.train.get_global_step(), 
                                               optimizer='Adagrad', 
                                               learning_rate=0.1)
    
    # 返回预测值\损失值和和优化步骤
    return predictions, loss, train_op
```

# 定义测试方法

```python
def run_eval(sess, test_X, test_label):
    # 将测试数据以数据集的形式提供给计算图
    ds = tf.data.Dataset.from_tensor_slices((test_X, test_label))
    ds = ds.batch(1)
    X, y = ds.make_one_shot_iterator().get_next()
    
    # 调用模型得到计算结果,这里不需要输入真实的label值
    with tf.variable_scope("model", reuse=True):
        prediction, _, _ = lstm_model(X, [0.0], False)
        
    # 将预测结果存入一个数组
    predictions = []
    labels = []
    for i in range(TESTING_EXAMPLES):
        p, l = sess.run([prediction, y])
        predictions.append(p)
        labels.append(l)
        
    # 计算rmse作为评价指标
    predictions = np.array(predictions).squeeze()
    labels = np.array(labels).squeeze()
    rmse = np.sqrt(((predictions - labels) ** 2).mean(axis=0))
    print("Root Mean Square Error is: %f" % rmse)
    
    # 对预测的sin函数曲线进行绘图
    plt.figure()
    plt.plot(predictions, label='predictions')
    plt.plot(labels, label='real_sin')
    plt.legend()
    plt.show()
```

# 执行训练和测试

```python
BATCH_SIZE = 32

# 将训练数据以数据集的方式提供给计算图
ds = tf.data.Dataset.from_tensor_slices((train_X, train_label))
ds = ds.repeat().shuffle(1000).batch(BATCH_SIZE)
X, y = ds.make_one_shot_iterator().get_next()

# 定义模型, 得到预测结果, 损失函数和训练操作
with tf.variable_scope("model"):
    _, loss, train_op = lstm_model(X, y, True)
    
# 定义计算
config = tf.ConfigProto(allow_soft_placement=True)
config.gpu_options.allow_growth = True
with tf.Session(config=config) as sess:
    # 变量初始化
    sess.run(tf.global_variables_initializer())
    
    print("X shape: ", sess.run(tf.shape(X)))
    print("y shape: ", sess.run(tf.shape(y)))
    
    # 测试在训练之前的模型效果
    print("Evaluate model before training ...")
    run_eval(sess, test_X, test_label)
    
    # 训练模型
    for i in range(TRAINING_EXAMPLES):
        _, l = sess.run([train_op, loss])
        if i % 1000 == 0:
            print("train step: " + str(i) + ", loss" + str(l))
            
    # 使用训练好的模型对数据进行预测
    print("Evaluate model after training.")
    run_eval(sess, test_X, test_label)
```

##### 3.1. os.walk

`os.walk`是一个简单易用的文件、目录遍历器。

* 函数原型：

```
walk(top, topdown=True, onerror=None, followlinks=False)
函数意义：
    os.walk 的返回值是一个生成器(generator),也就是说我们需要不断的遍历它，来获得所有的内。

参数：
    top:需要遍历的目录地址
    topdown:为True，则优先遍历top目录；否则，优先遍历top的子目录
    onerror:需要一个callabel对象，当walk异常时，会被调用
    followlinks:如果为真，则会遍历目录下的快捷方式(linux下是symbolic link)实际所指的目录(默认关闭)

返回值：
    每次遍历的对象都是返回的是一个三元组(root,dirs,files)：
    root：所指的是当前正在遍历的这个文件夹的本身的地址
    dirs：是一个 list ，内容是该文件夹中所有的目录的名字(不包括子目录)
    files：同样是 list , 内容是该文件夹中所有的文件(不包括子目录)
    
    如果topdown 参数为真，walk 会遍历top文件夹，与top文件夹中每一个子目录。
```

* 例子说明：

假设有如下文件结构：
```
a ->   b   ->   1.txt,  2.txt
             c   ->   3.txt
             d   ->   
           4.txt
           5.txt
```

遍历上述文件目录：
```
for (root, dirs, files) in os.walk('a'):
    # 第一次运行，当前遍历目录为a。
    # 三元组内容如下：
    root == 'a'
    dirs == ['b', 'c', 'd']
    files == ['4txt', '5.txt']
    
    # 接着遍历dirs中的每一个目录
    # 目录b
    root == 'a\\b'
    dirs == []
    files == ['1.txt', '2.txt']
    
    # 目录c
    root == 'a\\c'
    dirs == []
    files == ['3.txt']
    
    # 目录d
    root == 'a\\d'
    dirs == []
    files == []
    
    # 遍历完毕，退出循环
    
# 如果想获取文件的全部路径，只需要
for f in files:
    path = os.path.jion(root, f)

```

##### 3.3 glob.glob

glob模块是最简单的模块之一，内容非常少。用它可以查找符合特定规则的文件路径名。跟windows下的文件搜索差不多。
查找文件只用到三个匹配符：“*”，“？”，“[]”。"*"匹配0个或多个字符；“？”匹配单个字符；“[]”匹配指定范围内的字符，
如：[0-9]匹配数字

* glob.glob
返回所有匹配的文件路径列表。它只有一个参数pathname，定义了文件路径匹配规则，这里可以是绝对路径，也可以是相对路径。

* 例子说明
```
import glob

获取指定目录下的所有图片
print glob.glob("E:\pictures\*\*.jpg")

获取上级目录的所有.py文件
print glob.glob(r'../*.py') # 相对路径
```

##### 3.4 numpy.random.randint()

* 函数原型
```
numpy.random.randint(low, high=None, size=None, dtype)<br>
生成半开半闭区间[low, high)上离散均匀分布的整数值；若high=None，则取值区间变为[0, low)
```

* 例子说明
```
hight == None的情形<br>
a = np.random.randint(5, size=(2， 3))<br>
array([[2, 3, 2],<br>
       [2, 1, 4]])
       
hight != None<br>
a = np.random.randint(2, high=5, size=(2, 3))<br>
array([[2, 3, 4],<br>
       [3, 4, 4]])
```

````python
def create_image_lists(testing_percentage, validation_percentage):
    '''
    testing_percentage:指定了测试数据集的大小
    validation_percentage:指定了验证数据集的大小
    '''
    # 得到的所有图片都存在result这个字典里。
    # 这个字典的key为类别的名称，value也是一个字典，字典里存储了所有图片的名称。
    result = {}
    # 获取当前目录下所有的子目录:('INPUT_DATA', 
    # 'INPUT_DATA/daisy', 'INPUT_DATA/dandelion', 'INPUT_DATA/roses', 
    # 'INPUT_DATA/sunflowers', 'INPUT_DATA/tulips')
    sub_dirs = [x[0] for x in os.walk(INPUT_DATA)]
    # 得到的第一个目录是当前目录，不需要考虑
    is_root_dir = True
    for sub_dir in sub_dirs:
        if is_root_dir:
            is_root_dir = False
            continue
        
        # 获取当前目录下所有的有效图片文件夹
        extensions = ['jpg', 'jpeg', 'JPG', 'JPEG']
        file_list = [] # 元素为图片，图片格式为"INPUT_DATA/dir_name/图片名.图片后缀"
        # 获取目录名称
        dir_name = os.path.basename(sub_dir) # dir_name = daisy, dandelion, roses, sunflows, tulips
        for extension in extensions:
            file_glob = os.path.join(INPUT_DATA, dir_name, '*.' + extension) # 图片路径：INPUT_DATA/dir_name/*.图片后缀
            file_list.extend(glob.glob(file_glob)) # 获取路径INPUT_DATA/dir_name/*.图片后缀，所有图片
        if not file_list: 
            # 如果file_list为空，重新开始循环
            continue

        # 通过目录名获取类别的名称
        label_name = dir_name.lower()
        
        # 初始化当前类别的训练数据集、测试数据集和验证数据集
        # 列表中的元素格式为：“图片名.图片后缀”
        training_images = []
        testing_images = []
        validation_images = []
        for file_name in file_list:
            # 获取花名
            base_name = os.path.basename(file_name) # 获取图片名：“图片名.后缀”
            
            # 随机划分数据为训练数据集、测试数据集和验证数据集
            chance = np.random.randint(100)
            if chance < validation_percentage:
                validation_images.append(base_name)
            elif chance < (testing_percentage + validation_percentage):
                testing_images.append(base_name)
            else:
                training_images.append(base_name)

        # 将当前类别的数据放入结果字典中
        result[label_name] = { # label_name = daisy, dandelion, roses, sunflows, tulips
            'dir': dir_name, # 目录名称
            'training': training_images, 
            'testing': testing_images,
            'validation': validation_images,
            }
        
    # 返回整理好的所有数据
    return result
````

#### 4. 定义函数通过类别名称、所属数据集和图片编号获取一张图片的地址。

````python
def get_image_path(image_lists, image_dir, label_name, index, category):
    '''
    函数意义:
        根据特征向量根目录(image_dir)、图片种类(label_ame)、图片编号(index)和数据集类型(category)，
        创建该图片的特征向量地址
        
    参数意义:
        image_lists:列表，create_image_list函数的返回值
        image_dir：给出存放特征向量的根目录。存放图片数据的根目录和存放图片特征向量的根目录不同。
        label_name：花名，值可能为‘daisy’, ‘dandelion’, ‘roses’, ‘sunflows’, ‘tulips’
        index：一个随机数字，给定了需要获取的图片编号
        category:指定了需要获取的图片是在训练数据集、测试数据集还是验证数据集
        
    返回值:
        返回指定图片的特征向量地址
    '''
    # 获取给定的花的类别(label_name)中所有的图片信息
    label_lists = image_lists[label_name]
    # 获取指定花的类别的目录名称
    sub_dir = label_lists['dir'];
    
    # 根据所属数据集的名称获取该集合中全部图片信息
    category_list = label_lists[category]
    # 对随机数字进行取模:获取图片编号
    mod_index = index % len(category_list)
    
    # 根据图片编号(mod_index)获取图片的文件名(base_name)
    base_name = category_list[mod_index]
    
    # 最终的地址为数据根目录的地址加上类别的文件夹加上图片的名称
    full_path = os.path.join(image_dir, sub_dir, base_name)
    
    return full_path
````

#### 5. 定义函数获取Inception-v3模型处理之后的特征向量的文件地址。

````python
def get_bottleneck_path(image_lists, label_name, index, category):
    '''
    函数意义：
        提供花名(label_name)、数据集种类(category)和图片编号(index),计算特征向量的地址
    
    参数意义：
        image_lists:列表，create_image_list函数的返回值
        label_name:花名，值可能为‘daisy’, ‘dandelion’, ‘roses’, ‘sunflows’, ‘tulips’
        index:一个随机数字
        category:字符串，可能的取值为"training"、"testing"、"validation"
        
    返回值：
        返回指定图片的特征向量的地址
    '''
    return get_image_path(image_lists, CACHE_DIR, label_name, index, category) + '.txt'
````

#### 6. 定义函数使用加载的训练好的Inception-v3模型处理一张图片，得到这个图片的特征向量。

````python
def run_bottleneck_on_image(sess, image_data, image_data_tensor, bottleneck_tensor):
    '''
    函数意义：
        这个过程实际上就是讲图片作为输入计算瓶颈张量的值。
        这个瓶颈张量的值就是这张图片的新的特征向量。
    
    参数意义：
        sess:会话
        image_data:输入的数据
        image_data_tensor:输入的图片数据张量
        bottleneck_tensor:瓶颈层输出张量
        
    返回值:
        返回输入的图片数据经过inception-v3模型计算后的特征向量
    '''
    bottleneck_values = sess.run(bottleneck_tensor, {image_data_tensor: image_data})
    
    # 经过卷积神经网络处理的结构是一个四维数组，
    # 需要将这个结果压缩成一个特征向量(一维数组)
    # np.squeeze()只能移除所有的单维条目，并不能压缩成一个特征向量啊？
    bottleneck_values = np.squeeze(bottleneck_values)
    
    # 返回特征向量
    return bottleneck_values
````

#### 7. 定义函数会先试图寻找已经计算且保存下来的特征向量，如果找不到则先计算这个特征向量，然后保存到文件。
这个函数获取一张经过Inception-V3模型处理之后的特征向量。这个函数会优先试图寻找已经计算且保存下来的特征向量，
如果找不到则优先计算这个特征向量，然后保存到文件。

````python
def get_or_create_bottleneck(sess, image_lists, label_name, index, category, jpeg_data_tensor, bottleneck_tensor):
    '''
    函数意义：
        给定一张图片的花名(label_name)、编号(index)、数据集种类(category),获取给定图片的特征向量，如果该向量不存在，则重新
        计算该图片的特征向量。
        
    参数意义：
        sess:会话
        image_lists: 列表，create_image_list函数的返回值
        label_name: 花名，值可能为‘daisy’, ‘dandelion’, ‘roses’, ‘sunflows’, ‘tulips’
        index: 一个随机数字
        category：字符串，可能的取值为"training"、"testing"、"validation"
        jpeg_data_tensor：数据输入对应的张量
        bottleneck_tensor:瓶颈层对应的张量
        
    返回值：
        返回指定图片的特征向量
        
    '''
    # 获取一张图片对应的特征向量文件的路径
    # 获取指定花名的所有图片信息
    label_lists = image_lists[label_name]
    # 获取指定花名的目录名称
    sub_dir = label_lists['dir']
    # 判断指定花名的特征目录是否存在，如果不存在，创建该目录
    sub_dir_path = os.path.join(CACHE_DIR, sub_dir)
    if not os.path.exists(sub_dir_path): 
        os.makedirs(sub_dir_path)
    
    # 获取指定图片的特征向量路径
    bottleneck_path = get_bottleneck_path(image_lists, label_name, index, category)
    
    # 如果这个特征向量文件不存在，则通过Inception-V3模型来计算特征向量，并将计算的结果存入文件
    if not os.path.exists(bottleneck_path):
        # 获取原始的图片文件
        image_path = get_image_path(image_lists, INPUT_DATA, label_name, index, category)
        # 获取图片内容
        image_data = gfile.FastGFile(image_path, 'rb').read()
        # 通过Inception-V3模型计算特征向量
        bottleneck_values = run_bottleneck_on_image(sess, image_data, jpeg_data_tensor, bottleneck_tensor)
        # 将计算得到的特征向量存入文件
        bottleneck_string = ','.join(str(x) for x in bottleneck_values)
        with open(bottleneck_path, 'w') as bottleneck_file:
            bottleneck_file.write(bottleneck_string)
    else:
        # 直接从文件中获取图片相应的特征向量
        with open(bottleneck_path, 'r') as bottleneck_file:
            bottleneck_string = bottleneck_file.read()
        bottleneck_values = [float(x) for x in bottleneck_string.split(',')]
        
    # 返回得到指定图片的特征向量
    return bottleneck_values
````

#### 8. 这个函数随机获取一个batch的图片作为训练数据。

````python
def get_random_cached_bottlenecks(sess, n_classes, image_lists, how_many, category, jpeg_data_tensor, bottleneck_tensor):
    '''
    sess: 表示会话
    n_classes:数字，花的种类个数
    image_lists:列表，create_image_list函数的返回值
    how_many:BATCH
    category:字符串，可能的取值为"training"、"testing"、"validation"
    jpeg_data_tensor: 数据输入对应的张量
    bottleneck_tensor:瓶颈层对应的张量
    '''
    bottlenecks = []
    ground_truths = []
    for _ in range(how_many):
        # 随机一个类别和图片的编号加入当前的训练数据
        label_index = random.randrange(n_classes) # 产生[0,n_classes)之间的随机数，作为一个类别
        label_name = list(image_lists.keys())[label_index] # 获取一个类别的名字，值可能为daisy, dandelion, roses, sunflows, tulips
        image_index = random.randrange(65536) # 产生[0, 65536)之间的随机数
        bottleneck = get_or_create_bottleneck(
            sess, image_lists, label_name, image_index, category, jpeg_data_tensor, bottleneck_tensor)
        ground_truth = np.zeros(n_classes, dtype=np.float32)
        ground_truth[label_index] = 1.0
        bottlenecks.append(bottleneck)
        ground_truths.append(ground_truth)

    return bottlenecks, ground_truths
````

#### 9. 这个函数获取全部的测试数据，并计算正确率。
在最终测试的时候需要在所有的测试数据上计算正确率

````python
def get_test_bottlenecks(sess, image_lists, n_classes, jpeg_data_tensor, bottleneck_tensor):
    bottlenecks = []
    ground_truths = []
    label_name_list = list(image_lists.keys())
    # 枚举所有的类别和每个类别中的测试数据
    for label_index, label_name in enumerate(label_name_list):
        category = 'testing'
        for index, unused_base_name in enumerate(image_lists[label_name][category]):
            # 通过Inception-V3模型计算图片对应的特征向量，并将其加入最终数据列表
            bottleneck = get_or_create_bottleneck(sess, image_lists, label_name, index, category,jpeg_data_tensor, bottleneck_tensor)
            ground_truth = np.zeros(n_classes, dtype=np.float32)
            ground_truth[label_index] = 1.0
            bottlenecks.append(bottleneck)
            ground_truths.append(ground_truth)
    return bottlenecks, ground_truths
````

#### 10. 定义主函数。

````python
def main():
    # 读取所有图片
    image_lists = create_image_lists(TEST_PERCENTAGE, VALIDATION_PERCENTAGE)
    n_classes = len(image_lists.keys()) # 获取花的种类个数
    
    # 读取已经训练好的Inception-v3模型。
    # 谷歌已经将训练好的模型保存在GraphDefProtocol Buffer中，里面保存了每个节点取值的计算方法以及变量的取值。
    with gfile.FastGFile(os.path.join(MODEL_DIR, MODEL_FILE), 'rb') as f:
        graph_def = tf.GraphDef()
        graph_def.ParseFromString(f.read())
    
    # 加载读取的Inception-V3模型，
    # 并返回数据输入所对应的张量以及计算瓶颈层结果所对应的张量
    bottleneck_tensor, jpeg_data_tensor = tf.import_graph_def(
        graph_def, return_elements=[BOTTLENECK_TENSOR_NAME, JPEG_DATA_TENSOR_NAME])

    # 定义新的神经网络输入
    # 这个输入就是新的图片经过Inception-V3模型前向传播到达瓶颈层时的节点取值
    # 可以将这个过程类似的理解为一种特征提取。
    bottleneck_input = tf.placeholder(tf.float32, [None, BOTTLENECK_TENSOR_SIZE], name='BottleneckInputPlaceholder')
    # 定义新的标准答案输入
    ground_truth_input = tf.placeholder(tf.float32, [None, n_classes], name='GroundTruthInput')
    
    # 定义一层全链接层来解决新的图片分类问题
    # 因为训练好的Inception-V3模型已经将原始的图片抽象为了更加容易分类的特征向量，
    # 所以不再需要训练那么复杂的神经网络来完成这个新的分类任务
    with tf.name_scope('final_training_ops'):
        weights = tf.Variable(tf.truncated_normal([BOTTLENECK_TENSOR_SIZE, n_classes], stddev=0.001))
        biases = tf.Variable(tf.zeros([n_classes]))
        logits = tf.matmul(bottleneck_input, weights) + biases
        final_tensor = tf.nn.softmax(logits)
        
    # 定义交叉熵损失函数。
    cross_entropy = tf.nn.softmax_cross_entropy_with_logits(logits=logits, labels=ground_truth_input)
    cross_entropy_mean = tf.reduce_mean(cross_entropy)
    train_step = tf.train.GradientDescentOptimizer(LEARNING_RATE).minimize(cross_entropy_mean)
    
    # 计算正确率。
    with tf.name_scope('evaluation'):
        correct_prediction = tf.equal(tf.argmax(final_tensor, 1), tf.argmax(ground_truth_input, 1))
        evaluation_step = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))

    with tf.Session() as sess:
        # 执行变量初始化的操作
        init = tf.global_variables_initializer()
        sess.run(init)
        
        # 训练过程。
        for i in range(STEPS):
            # 每次获取一个batch的训练数据
            train_bottlenecks, train_ground_truth = get_random_cached_bottlenecks(
                sess, n_classes, image_lists, BATCH, 'training', jpeg_data_tensor, bottleneck_tensor)
            sess.run(train_step, feed_dict={bottleneck_input: train_bottlenecks, ground_truth_input: train_ground_truth})
            
            # 在验证数据集上测试正确率
            if i % 100 == 0 or i + 1 == STEPS:
                validation_bottlenecks, validation_ground_truth = get_random_cached_bottlenecks(
                    sess, n_classes, image_lists, BATCH, 'validation', jpeg_data_tensor, bottleneck_tensor)
                validation_accuracy = sess.run(evaluation_step, feed_dict={
                    bottleneck_input: validation_bottlenecks, ground_truth_input: validation_ground_truth})
                print('Step %d: Validation accuracy on random sampled %d examples = %.1f%%' %
                    (i, BATCH, validation_accuracy * 100))
            
        # 在最后的测试数据上测试正确率。
        test_bottlenecks, test_ground_truth = get_test_bottlenecks(
            sess, image_lists, n_classes, jpeg_data_tensor, bottleneck_tensor)
        test_accuracy = sess.run(evaluation_step, feed_dict={
            bottleneck_input: test_bottlenecks, ground_truth_input: test_ground_truth})
        print('Final test accuracy = %.1f%%' % (test_accuracy * 100))

if __name__ == '__main__':
    main()
````

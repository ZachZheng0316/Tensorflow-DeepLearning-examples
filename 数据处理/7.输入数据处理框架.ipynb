# 输入数据处理框架




1. 获取存储数据的文件列表;
2. 生成并维护一个输入文件队列;
3. 从文件中读取数据;
4. 数据预处理;
5. 整理成batch作为模型的输入

```python
import tensorflow as tf
```

```python
# 1.获取存储文件列表
files = tf.train.match_filenames_once("../../../other/test/data.tfrecords-*")

# 2.创建输入文件队列
filename_queue = tf.train.string_input_producer(files, shuffle=False)

# 3.从文件中读取数据
reader = tf.TFRecordReader()

_, serialzed_example = reader.read(filename_queue)

feature_i = tf.FixedLenFeature([], tf.int64)
feature_j = tf.FixedLenFeature([], tf.int64)
features_map = {'i': feature_i, 'j': feature_j}
features = tf.parse_single_example(serialzed_example, features=features_map)

example, label = features['i'], features['j']

# 4.数据预处理
example_new = example * 100

# 5.把数据整理成batch
batch_size = 3
capacity = 1000 + 3 * batch_size
min_after_dequeue = 30

example_batch, label_batch = tf.train.shuffle_batch(
    [example_new, label], 
    batch_size=batch_size, 
    capacity=capacity, 
    min_after_dequeue=min_after_dequeue)

# 执行
with tf.Session() as sess:
    sess.run([tf.global_variables_initializer(), tf.local_variables_initializer()])
    
    # 打印文件列表
    print("files: ", sess.run(files))
    
    coord = tf.train.Coordinator()
    threads = tf.train.start_queue_runners(sess=sess, coord=coord)
    
    # 获取打印之后的列表; 在真实的问题中，这输出一般会作为神经网络输入
    for i in range(2):
        cur_example_batch, cur_label_batch = sess.run([example_batch, label_batch])
        print(cur_example_batch, cur_label_batch)
        
    coord.request_stop()
    coord.join(threads)
```
